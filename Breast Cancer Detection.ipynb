# Breast Cancer Detection

```python
# Workflow of Machine Learning Model(Bicycle)
'''
0) Import Important dependencies
1) Setting up the data
2) Creating the Model
3) Training the Model
4) Testing the Model
5) Prediction
'''
```

## 1. Import libraries

```python
# Import Important dependencies
import numpy as np
import pandas as pd
```

```python
# Importing Logistic Regression Model (Binary Outputs)
from sklearn.linear_model import LogisticRegression
```

```python
# Importing train test split function
from sklearn.model_selection import train_test_split
```

```python
# Import accuracy function
from sklearn.metrics import accuracy_score
```

## 2. Load Dataset

```python
# Importing the dataset
df = pd.read_csv("breast-cancer.csv")
df
```

```python
# Two types of tumors
# Benign - Treatable tumor
# Malignant - Non-Treatable tumor
```

```python
# Check number of Beningn and Malignant Tumor
df['diagnosis'].value_counts().reset_index()
```

## 3. Data preprocessing

```python
# Checking the info
df.info()
```

```python
# Checking the null values
df.isnull().sum()
```

```python
# Removing unwanted columns
df.drop(['id'], axis = 1, inplace = True)
```

## 4. Label Encoding

```python
# Convert B-1 and M-0 in the diagnosis column
df.loc[df['diagnosis'] == 'B', 'diagnosis'] = 1
df.loc[df['diagnosis'] == 'M', 'diagnosis'] = 0
```

```python
# Split the data into Input Feature - X
# Target Column - Y
X = df.drop(['diagnosis'], axis = 1)
Y = df['diagnosis']
```

```python
X
```

```python
Y
```

## 5. Train test split

```python
# Split the data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 42)
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
# Convert Y_train and Y_test as integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

## 6. Model Training

```python
# Extract Model
model = LogisticRegression(max_iter = 10000)
# Training the ML model using training data
model.fit(X_train, Y_train)
```

## 7. Evaluate model accuracy

```python
# Checking accuracy score
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print("Accuracy score of training data is",training_data_accuracy)
```

```python
# Checking accuracy score
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print("Accuracy score of testing data is",testing_data_accuracy)
```
